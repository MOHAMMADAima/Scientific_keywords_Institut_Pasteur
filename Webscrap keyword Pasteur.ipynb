##### Extracting keywords for each section :
###### 1. DOMAINES SCIENTIFIQUES

```python
# Extraction keywords of section DOMAINES SCIENTIFIQUES

def keywords_DS(b):
    DS=[]
    tag=b.findAll("li", class_="tag scientific_field")
    for t in (tag):
        DS.append(t.text)
        
    return DS
keywords_DS(bs)
```

```python
# Extraction keywords of section ORGANISME ET ORGANE

def keywords_OO(b):
    OO=[]
    tag=b.findAll("li", class_="tag organism")
    for t in (tag):
        OO.append(t.text)
    return OO
keywords_OO(bs)
```

```python
# Extraction keywords of section APPLICATION
def keywords_AP(b):
    AP=[]
    tag=b.findAll("li", class_="tag application")
    for t in (tag):
        AP.append(t.text)
    return AP

keywords_AP(bs)
```

```python
# Extraction keywords of section TECHNIQUE

def keywords_TE(b):
    TE=[]
    tag=b.findAll("li", class_="tag technique")
    for t in (tag):
        TE.append(t.text)
    return TE
keywords_TE(bs)
```

```python
# Extraction keywords of section MALADIE
def keywords_MA(b):
    MA=[]
    tag=b.findAll("li", class_="tag disease")
    for t in (tag):
        MA.append(t.text)
    return MA
keywords_MA(bs)
```

```python
## Function : input ->  a department weblink | output -> table with keywords per units of the department

def keywords_dep(dept_url):
    request = requests.get(dept_url) 
    bs = BeautifulSoup(request.text)
    

    unit_nb=str(bs.findAll("div", class_="s-text")).split("<") # Number of units in the department
    for i in unit_nb:    
        if "Équipes" in i:
            nb_of_unit=int(re.findall('"s-text">(.*) Équipes',i)[0])
            print(f'There are {nb_of_unit} research units in this departments.')
            break
     # Note : Here are the number of units per departments (same others as in list DEPT) : [17, 8, 17, 12, 13, 13, 15, 20, 6, 10, 9, 19, 23]
        
    dep_name=max(str(bs.find("h1")).split("<"),key=len)[3:] # Name of the current department
    unit_name=str(bs.findAll("a", class_="team")).split("<") # Names of the all research units of the current page
    
    
    UN=[] # Collect the weblink of research units
    n=0
    for i in range (len(unit_name)):
        if n < nb_of_unit:  
            if 'class="team"' in unit_name[i]:
                href=re.findall('href="(.*)" rel',unit_name[i])[0]
                UN.append(href)
                n+=1    
        else: 
            break
    
    df=pd.DataFrame(columns=['Department',"Research Units","DOMAINES SCIENTIFIQUES","MALADIES","ORGANISMES ET ORGANES","APPLICATION","TECHNIQUES"])
   

    for i in range (len(UN)): # Loop on research units' webpage to collect keywords
        actions = ActionChains(driver)
        driver.get(UN[i])
        driver.find_element(by=By.XPATH, value="/html/body/div[1]/div[10]/div[8]/div[4]/div/div[1]/div[1]/div[2]/a").click()
        get_url = driver.current_url
        request = requests.get(get_url)
        bss = BeautifulSoup(request.text) 
        time.sleep(3)

        un_name=max(str(bss.find("h1")).split("<"),key=len)[3:] # Name of the research unit of current webpage
        
        # Collecting keywords per section of the current research units
        DS=keywords_DS(bss)
        MA=keywords_MA(bss)
        OO=keywords_OO(bss)
        AP=keywords_AP(bss)
        TE=keywords_TE(bss)

       # Create dataframe of current research unit and its keywords
        df1=pd.DataFrame([[ dep_name, un_name,keywords_DS(bss),keywords_MA(bss),keywords_OO(bss)
                           ,keywords_AP(bss),keywords_TE(bss)]],
                                    columns=['Département',"Unité","DOMAINES SCIENTIFIQUES","MALADIES",
                                             "ORGANISMES ET ORGANES","APPLICATION","TECHNIQUES"])
        
        # Concatenate all dataframes of each research units of a departments and their keywords
        df=pd.concat([df,df1])
                
                
    return(df)  
```
